```python
x = np.linspace(0, np.pi*4)
```

```python
y = np.sin(x)
```

```python
pyplot.plot(x,y)
```

```python
x, y = np.meshgrid(x, x)
```

```python
z = x **2 + y**2
```

```python
pyplot.imshow(z)
pyplot.colorbar()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = pyplot.figure()
axes = fig.gca(projection='3d')
surf = axes.plot_surface(x,y,z, cmap='viridis')
```
